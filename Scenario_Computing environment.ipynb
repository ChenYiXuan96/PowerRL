```python

# 定义成本函数
def compute_gen_cost(net):
    """
    This function computes the total generation cost.
    ***Parameter***:
    poly_cost: net.poly_cost
    res_gen: net.res_gen
    ***return***:
    cost: The first element is ext_grid_cost, The left are gen_cost.
    """
    
    res1 = net.poly_cost[net.poly_cost['et'] == 'gen'][['element', 'cp2_eur_per_mw2']].astype({'element': int}).set_index('element')['cp2_eur_per_mw2'] * net.res_gen['p_mw'] ** 2 + \
    net.poly_cost[net.poly_cost['et'] == 'gen'][['element', 'cp1_eur_per_mw']].astype({'element': int}).set_index('element')['cp1_eur_per_mw'] * net.res_gen['p_mw'] + \
    net.poly_cost[net.poly_cost['et'] == 'gen'][['element', 'cp0_eur']].astype({'element': int}).set_index('element')['cp0_eur']
    
    res2 = net.poly_cost[net.poly_cost['et'] == 'ext_grid'][['element', 'cp2_eur_per_mw2']].astype({'element': int}).set_index('element')['cp2_eur_per_mw2'] * net.res_ext_grid['p_mw'] ** 2 + \
    net.poly_cost[net.poly_cost['et'] == 'ext_grid'][['element', 'cp1_eur_per_mw']].astype({'element': int}).set_index('element')['cp1_eur_per_mw'] * net.res_ext_grid['p_mw']
    
    if net.res_ext_grid['p_mw'].values[0] < 0:
        print('The external grid absorbs power.')
        res2.iloc[0] = 0
        
    return res2.append(res1)
    

compute_gen_cost(net)
```

```python
# 定义绘图函数
def plot_with_DER(net):
    fig, ax = plt.subplots(figsize=[12,8])
    ax.scatter(net.bus_geodata.loc[net.gen['bus']]['x'], net.bus_geodata.loc[net.gen['bus']]['y']+0.15, s=200, c='r', marker='*')
    simple_plot(net, plot_loads=True, ax=ax)
    ax.legend()
plot_with_DER(net)
```

```python
# Randomly add DERs...

WIND_CAPACITY = 200
PV_CAPACITY = 15
wind_num = random.choice([1, 2])
pv_num = random.choice(range(2,9))
wind_poses = np.random.choice(net.bus.index.tolist(), size=wind_num, replace=False)
pv_poses = np.random.choice(net.bus.index.tolist(), size=pv_num, replace=False)

for wind_pos in wind_poses:
    pp.create_sgen(net, wind_pos, max(0, WIND_CAPACITY + np.random.normal(scale=50)), type='Wind_MV')
for pv_pos in pv_poses:
    pp.create_sgen(net, pv_pos, max(0, PV_CAPACITY + np.random.normal(scale=5)), type='PV_MV')
    
# Allocate profiles for DERs...

valid_wind_names = []
valid_pv_names = []
for name in net.profiles['renewables'].columns.to_list():
    if name[:2] == 'PV':
        valid_pv_names.append(name)
    elif name[:2] == 'WP':
        valid_wind_names.append(name)
        
wind_profiles = random.choices(valid_wind_names, k=wind_num)
pv_profiles = random.choices(valid_pv_names, k=pv_num)
net.sgen.loc[net.sgen['type'] == 'Wind_MV', 'profile'] = wind_profiles
net.sgen.loc[net.sgen['type'] == 'PV_MV', 'profile'] = pv_profiles

# Randomly add redispatchable generations...

NEW_GEN_CAPACITY = 25
new_gen_num = random.choice(range(2, 6))
available_poses = list(set(net.bus.index.tolist()) - set(net.gen.bus.tolist()))
new_gen_poses = np.random.choice(available_poses, size=new_gen_num, replace=False)
for gen_pos in new_gen_poses:
    maximum_gen = max(0, NEW_GEN_CAPACITY + np.random.normal(scale=5))
    pp.create_gen(net, gen_pos, maximum_gen, max_p_mw=maximum_gen, min_p_mw=0)
    current_idx = max(net.gen.index)
    pp.create_poly_cost(net, current_idx, et='gen', cp1_eur_per_mw=2.5, cp2_eur_per_mw2=0.0175)
```

```python
net.sgen.loc[net.sgen['type'] == 'Wind_MV', 'profile'] = wind_profiles
net.sgen.loc[net.sgen['type'] == 'PV_MV', 'profile'] = pv_profiles
```
